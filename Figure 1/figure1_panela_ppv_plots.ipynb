```python
import pandas as pd
import numpy as np
import sklearn.ensemble
import sklearn.model_selection
import matplotlib.pyplot as plt
import statsmodels.stats.proportion
import scipy.special
plt.rcParams['pdf.use14corefonts'] = True
```

```python
guardant_table = pd.read_excel('../data/Leshchiner_et_al_Tables.xlsx', sheet_name='SuppTable2_GuardantDatabase', index_col=0)
guardant_drug_table = pd.read_csv('../data/treatment_information_for_classifier.tsv', sep='\t', index_col=0)
```

```python
# Get list of genes in panel
gene_list = list(pd.unique(guardant_table['gene']))
```

```python
# List of drugs
drug_index = ['kadcyla', 'tdm-1', 'tykerb', 'nerlynx', 'palbociclib', 'ibrance', 'fulvestrant', 'faslodex', 'ribociclib', 'kisqali', 'tamoxifen',
              'abemiciclib', 'verzenio', 'letrozole', 'femara', 'exemestane', 'aromasin', 'anastrozole', 'arimidex', 'anti-her2']
```

```python
annot_mat = pd.read_excel('../data/Leshchiner_et_al_Tables.xlsx', sheet_name='SuppTable4_ClassifierTraining', index_col=0).iloc[:, :-2]
er_annot_ser = pd.read_csv('../data/classifier_er_labels.tsv', sep='\t', index_col=0)['0']
her2_annot_ser = pd.read_csv('../data/classifier_her2_labels.tsv', sep='\t', index_col=0)['0']
```

```python
patients = list(annot_mat.index)
```

```python
# Set random seed
np.random.seed(999999)
```

```python
# Train classifier on ER labels
er_cols = list(annot_mat.columns)
er_cols.remove('anti-her2') # Avoid overfitting on known HER2 features
er_cols.remove('ERBB2_amp')

er_feature_mat = annot_mat[er_cols]

kfold = sklearn.model_selection.KFold(n_splits=10, shuffle=True)
er_predictions = np.zeros((len(er_annot_ser), 2))
er_classifiers = []
feature_importances = {eve: [] for eve in er_feature_mat.columns}
for i, (train, test) in enumerate(kfold.split(er_annot_ser.index)):
    classifier = sklearn.ensemble.RandomForestClassifier()
    classifier.fit(er_feature_mat.iloc[train], er_annot_ser.iloc[train])
    er_classifiers.append(classifier)
    predictions = classifier.predict_proba(er_feature_mat.iloc[test])
    er_predictions[test, :] = predictions
    feature_importance = classifier.feature_importances_
    for eve, f in zip(er_feature_mat.columns, feature_importance):
        feature_importances[eve].append(f)
```

```python
# Train classifiers on HER2 labels
er_pos_patients = list(er_feature_mat.loc[er_annot_ser].index)
er_neg_patients = list(er_feature_mat.loc[~er_annot_ser].index)
her2_annot_er_pos = her2_annot_ser.loc[er_pos_patients]
her2_annot_er_neg = her2_annot_ser.loc[er_neg_patients]

kfold = sklearn.model_selection.KFold(n_splits=10, shuffle=True)
her2_predictions_er_pos = np.zeros((len(er_pos_patients), 2))
her2_classifiers_er_pos = []
er_pos_feature_mat = annot_mat.loc[er_pos_patients]
feature_importances_er_pos = {eve: [] for eve in annot_mat.columns}
for i, (train, test) in enumerate(kfold.split(er_pos_patients)):
    classifier = sklearn.ensemble.RandomForestClassifier()
    classifier.fit(er_pos_feature_mat.iloc[train], her2_annot_er_pos.iloc[train])
    her2_classifiers_er_pos.append(classifier)
    predictions = classifier.predict_proba(er_pos_feature_mat.iloc[test])
    her2_predictions_er_pos[test, :] = predictions
    for eve, f in zip(annot_mat.columns, classifier.feature_importances_):
        feature_importances_er_pos[eve].append(f)

kfold = sklearn.model_selection.KFold(n_splits=10, shuffle=True)
her2_predictions_er_neg = np.zeros((len(er_neg_patients), 2))
her2_classifiers_er_neg = []
er_neg_feature_mat = annot_mat.loc[er_neg_patients]
feature_importances_er_neg = {eve: [] for eve in annot_mat.columns}
for i, (train, test) in enumerate(kfold.split(er_neg_patients)):
    classifier = sklearn.ensemble.RandomForestClassifier()
    classifier.fit(er_neg_feature_mat.iloc[train], her2_annot_er_neg.iloc[train])
    her2_classifiers_er_neg.append(classifier)
    predictions = classifier.predict_proba(er_neg_feature_mat.iloc[test])
    her2_predictions_er_neg[test, :] = predictions
    for eve, f in zip(annot_mat.columns, classifier.feature_importances_):
        feature_importances_er_neg[eve].append(f)
```

```python
# Function for aggregating predictions of cross-validation classifiers

def agg_predict(arr, classifiers, n_classes=2):
    predictions = np.zeros((len(arr), n_classes))
    for classifier in classifiers:
        predictions += np.log(classifier.predict_proba(arr))
    predictions /= 10
    predictions = np.exp(predictions - scipy.special.logsumexp(predictions, axis=1, keepdims=True))
    return predictions
```

```python
# Create full event matrix of guardant genomic data

guardant_mut_table = guardant_table.loc[(guardant_table['mut_type'] != 'synonymous') & (guardant_table['mut_type'] != 'amp')
  ].groupby(['patient_id', 'gene']).size().unstack().fillna(0).astype(bool).astype(int)
guardant_amp_table = guardant_table.loc[(guardant_table['mut_type'] == 'amp')
  ].groupby(['patient_id', 'gene']).size().unstack().fillna(0).astype(bool).astype(int)
for gene in gene_list:
    if gene not in guardant_mut_table.columns:
        guardant_mut_table[gene] = 0
    if gene not in guardant_amp_table.columns:
        guardant_amp_table[gene] = 0
guardant_mut_table = guardant_mut_table[gene_list]
guardant_amp_table = guardant_amp_table[gene_list]
guardant_amp_table.columns += '_amp'

full_mat = pd.concat([guardant_mut_table, guardant_amp_table, guardant_drug_table], axis=1).fillna(0)[annot_mat.columns]
```

```python
# Run classifiers on unannotated guardant samples

features_full_mat = full_mat.loc[~full_mat.index.isin(patients)]
full_er_predictions = agg_predict(features_full_mat[er_cols], er_classifiers)
full_her2_er_pos_given_predictions = agg_predict(features_full_mat, her2_classifiers_er_pos)
full_her2_er_neg_given_predictions = agg_predict(features_full_mat, her2_classifiers_er_neg)

classifier_results = pd.DataFrame({'p(ER+)': {s: p for s, p in zip(features_full_mat.index, full_er_predictions[:, 1])},
 'p(HER2+|ER+)': {s: p for s, p in zip(features_full_mat.index, full_her2_er_pos_given_predictions[:, 1])},
 'p(HER2+|ER-)': {s: p for s, p in zip(features_full_mat.index, full_her2_er_neg_given_predictions[:, 1])}})
classifier_results.index.name = 'patient'
# classifier_results.to_csv('../data/classifier_results.tsv', sep='\t')
```

```python
# ER classifier PPV plot
er_prob_vals = np.zeros(81)
er_prob_vals_low = np.zeros(81)
er_prob_vals_high = np.zeros(81)
pos_predictions = er_predictions[er_annot_ser, 1]
neg_predictions = er_predictions[~er_annot_ser, 1]
n_samples = np.zeros(81)
for i, x in enumerate(np.arange(.1, .91, .01)):
    pos = len(pos_predictions[(pos_predictions > x - .1) & (pos_predictions < x + .1)])
    neg = len(neg_predictions[(neg_predictions > x - .1) & (neg_predictions < x + .1)])
    er_prob_vals[i] = pos / (pos + neg)
    er_prob_vals_low[i], er_prob_vals_high[i] = statsmodels.stats.proportion.proportion_confint(pos, pos + neg, .9, method='beta')
    n_samples[i] = pos + neg
fig, axes = plt.subplots(figsize=(3, 7), nrows=2, gridspec_kw={'height_ratios':(6, 1)}, sharex=True)
axes[0].scatter(np.arange(.1, .91, .01), er_prob_vals, s=5, c='g')
axes[0].vlines(np.arange(.1, .91, .01), er_prob_vals_low, er_prob_vals_high, lw=.5, color='k', zorder=-1)
axes[0].set_xlim(0, 1)
axes[0].set_ylim(0, 1)
axes[0].set_title('ER classifier PPV', fontsize=18)
axes[0].set_ylabel('Frequency ER+', fontsize=14)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['left'].set_visible(False)

axes[1].bar(np.arange(.1, .91, .01), n_samples, width=.01)
axes[1].set_xlabel('Predicted probability ER+', fontsize=14)
axes[1].set_ylabel('# samples', fontsize=14)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['bottom'].set_visible(False)
axes[1].spines['left'].set_visible(False)
# plt.savefig('ER_classifier_ppv_v4.pdf', format='pdf', bbox_inches='tight')
plt.show()
plt.close()
```

```python
def confusion_mat(predictions, annot_ser, thresh):
    return np.array([[annot_ser[predictions > thresh].sum(), annot_ser[predictions <= thresh].sum()],
                     [(~annot_ser[predictions > thresh]).sum(), (~annot_ser[predictions <= thresh]).sum()]])

plt.figure()
ax = plt.gca()
mat = confusion_mat(er_predictions[:, 0] * full_her2_predctions_er_neg[:, 0], ~(er_annot_ser | her2_annot_ser), .5)
ax.matshow(mat)
for (i, j), z in np.ndenumerate(mat):
    ax.text(j, i, '{:d}'.format(z), ha='center', va='center', color='white' if z < np.sum(mat) * .15 else 'black')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Predicted ER-/HER2-', 'Not predicted ER-/HER2-'])
ax.set_yticks([0, 1])
ax.set_yticklabels(['Annotated ER-/HER2-', 'Not annotated ER-/HER2-'])
plt.show()
plt.close()
```
