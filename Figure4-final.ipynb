```python
#CLONALITY PLOTS
mergedr=merged[merged.mut_type!="synonymous"]
pivot_sublones=pd.pivot_table(mergedr, values='af_ratio', columns=["test_num"],
                    index=['patient_id','gene',"alteration"], aggfunc=np.max)
pivot_sublones
```

```python
sel=pivot_sublones
sel[sel.index.get_level_values("gene")=="ESR1"].iloc[:,0:4]
sel=sel[sel.index.get_level_values("gene")=="ESR1"].iloc[:,0:4]
sel=sel[list((sel.T.loc[1].notna()))&(sel.T.loc[2].notna())]
ax=sns.lineplot(data=sel.T, lw=1, alpha=0.8,errorbar=None,legend=False, dashes=False, palette=sns.color_palette("light:seagreen",256))
ax.set_ylim(0,1)
#ax.get_legend().remove()
ax.set_xticks([1,2,3,4])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Time Points')
ax.set_ylabel('ESR1 Relative AF')

#ax.set_labels('ESR1 Relative AF', 'Timepoint')

plt.savefig("ESR1_clonality.pdf")
```
